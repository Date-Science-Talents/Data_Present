### Clean df

```python
df1 = df.drop(columns=['VisitorID', 'Users', 'Accommodation', 'ContentInfo',
                            'DateOfBirth', 'StaffPlanConditions']).applymap(str)
df1
```

```python
df1 = df1[['Time', 'ExternalID', 'Gender', 'Age', 'CareSysCondition', 'LinkTitle', 'LinkType']]
df1
```

```python
df1['Gender'].unique()
```

```python
df_clean = df1[~df1.Gender.str.contains("U")]
df_clean
```

```python
df_clean['Gender'].unique()
```

```python
df_clean['Age'].unique()
```

### Get Top20 Activity

```python
df_clean.groupby(['LinkTitle']).count().sort_values(by='Time', ascending=False).head(20)
```

```python
value_list = ['Entertainment', 'Card Matching Games', 'YouTube', 'My Music', 'My Interests',
             'Google', 'Patience', 'Play Games', 'Videos', 'Online Tv',
             'Information', 'Youtube', 'Radio Stations', 'Single Player Games', 'Family & Friends',
             'Smooth Radio', 'Yahoo Mail', 'Test Cards', 'Youtube Music', 'Candy Crush Saga']
df_t20 = df_clean[df_clean.LinkTitle.isin(value_list)]
df_t20
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot( 'CareSysCondition', 'LinkTitle', data=df_t20, linestyle='', marker='o')
plt.xlabel('Value of CareSysCondition')
plt.ylabel('Value of LinkTitle')
plt.title('Overplotting looks like that:', loc='left')
```

```python
plt.plot( 'Gender', 'LinkTitle', data=df_t20, linestyle='', marker='o')
plt.xlabel('Value of Gender')
plt.ylabel('Value of LinkTitle')
plt.title('Overplotting looks like that:', loc='left')
```
