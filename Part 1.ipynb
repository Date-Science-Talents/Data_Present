### Data cleaning

```python
df_clean = df.drop(columns=['VisitorID', 'Users', 'Accommodation', 'ContentInfo']).applymap(str)
df_clean
```

### Most popular activities in general

```python
df_clean.groupby(['LinkTitle']).count().sort_values(by='Time', ascending=False).head(20)
```

### Most popular built in activities

```python
df_bi = df_clean[df_clean.LinkType.str.contains('Category')]
df_bi
```

```python
df_bi.groupby(['LinkTitle']).count().sort_values(by='Time', ascending=False).head(20)
```

### Most popular Internet activities

```python
df_it = df_clean[df_clean.LinkType.str.contains('Internet')]
df_it
```

```python
df_it.groupby(['LinkTitle']).count().sort_values(by='Time', ascending=False).head(20)
```
